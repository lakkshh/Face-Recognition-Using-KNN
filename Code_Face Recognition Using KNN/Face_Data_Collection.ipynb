```python
import numpy as np
import cv2
```

```python
cap=cv2.VideoCapture(0)
face_cascade=cv2.CascadeClassifier("C:/Users/surender/Face Detect/Face Recognition using Haar Cascade Classifier/haarcascade_frontalface_alt.xml")
path='./face_data/'
file_name=input("Enter your name ")
face_data=[]
```

```python
while True:
    ret,frame=cap.read()
    
    if(ret==False):
        continue

    #frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(frame,1.3,5)
    
    if(len(faces)==0):
        continue
        
    faces=sorted(faces, key=lambda f:f[2]*f[3])
    
    for face in faces[-1:]:
        x,y,w,h=face
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),3)
        offset=10
        face_section=frame[y-offset:y+offset+h,x-offset:x+offset+w]
        face_section=cv2.resize(face_section,(100,100))
        face_data.append(face_section)
        
        '''eye = eye_cascade.detectMultiScale(frame)
        
        if(len(eyes)==0):
            continue
            
            x2,y2,w2,h2=eye
            cv2.rectangle(frame,(x2,y2),(x2+w2,y2+h2),(0,0,255),3)'''
            
    
    cv2.imshow("frame",frame)
    
    key_press=cv2.waitKey(1) & 0XFF
    if(key_press==ord('q')):
        break

face_data=np.asarray(face_data)
print(face_data)
print(face_data.shape)
    
face_data=face_data.reshape((face_data.shape[0],-1))
print(face_data)
print(face_data.shape)

np.save(path+file_name+'.npy',face_data)
print("Data Saved")
        
cap.release()
cv2.destroyAllWindows()
```
